**1. Importing Libraries**

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier #import decision tree function
from sklearn.model_selection import train_test_split #import train-test function
from sklearn import metrics #accuracy calculations
import matplotlib.pyplot as plt #visualize the tree
from sklearn import tree 
```

**2. Loading the Dataset**

```python
data = pd.read_csv("heart_Disease.csv")
data.head()
```

**3. Splitting into features and target attribute**

```python
feature_cols = ['age', 'sex', 'trestbps', 'chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']
X = data[feature_cols] # Features
y = data.target # Target variable
```

**4. Train and test split where 70 percent of the data is training data**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

**5. Creating the decision tree**

A decision tree is a flowchart-like tree structure where an internal node represents feature(or attribute), the branch represents a decision rule, and each leaf node represents the outcome. The topmost node in a decision tree is known as the root node. It learns to partition on the basis of the attribute value. It partitions the tree in recursively manner call recursive partitioning. This flowchart-like structure helps you in decision making. It's visualization like a flowchart diagram which easily mimics the human level thinking. That is why decision trees are easy to understand and interpret.

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

**6. Printing the accuracy**

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

**7. Visualizing the tree**

```python
fig, ax = plt.subplots(figsize=(10, 10))
tree.plot_tree(clf, fontsize=10)
plt.show()
```
